## DATA PROCESSING

### Get Places ID

```python
df_places = pd.read_excel('indonesia_tourism (3).xlsx', sheet_name="Sheet3")
```

```python
places_id = []
for place in df_places.columns:
  try:
    place_json = gmaps.places(place)
    if place_json['results']:  # Check if there are results
      place_id = place_json['results'][0]['place_id']
      places_id.append(place_id)
  except (KeyError, IndexError):
    # Handle cases where 'results' key is missing or empty
    pass
```

```python
places_data = []
```

### Get The Review

```python
max_reviews = 5
default_review = 'blank'
default_rating = 0

places_data = []

for place_id, place_name in zip(places_id, df_places.columns):
    place_json = gmaps.place(place_id=place_id)
    result = place_json.get('result', {})
    place_data = {'id': place_id, 'name': place_name, 'reviews': [], 'ratings': []}
    
    # Check if 'reviews' key exists and is non-empty
    if 'reviews' in result and len(result['reviews']) > 0:
        # Iterate over each review up to max_reviews
        for review in result['reviews'][:max_reviews]:
            text = review.get('text', default_review)
            rating = review.get('rating', default_rating)
            place_data['reviews'].append(text)
            place_data['ratings'].append(rating)
        
        # If there are fewer reviews than max_reviews, pad with default values
        while len(place_data['reviews']) < max_reviews:
            place_data['reviews'].append(default_review)
            place_data['ratings'].append(default_rating)
    else:
        # Handle case where there are no reviews
        place_data['reviews'] = [default_review] * max_reviews
        place_data['ratings'] = [default_rating] * max_reviews
    
    # Append the place_data dictionary after processing all reviews
    places_data.append(place_data)

# Example to print the results
for data in places_data:
    print(f"Place ID: {data['id']}")
    print(f"Name: {data['name']}")
    print(f"Reviews: {data['reviews']}")
    print(f"Ratings: {data['ratings']}")
    print("-" * 20)
```

## DATA STORAGE

```python
import pandas as pd

# Step 1: Convert places_data to a dictionary
dict_places = {place['name']: place['reviews'] for place in places_data}

# Define max_reviews if needed to pad reviews to this length
max_reviews = 5
default_review = 'blank'

# Step 2: Ensure all places have the same number of reviews
for place in dict_places:
    while len(dict_places[place]) < max_reviews:
        dict_places[place].append(default_review)

# Step 3: Create a DataFrame
# Initialize an empty dictionary to store data for the DataFrame
df_data = {place: [] for place in dict_places}

# Fill the DataFrame data
for i in range(max_reviews):
    for place in dict_places:
        df_data[place].append(dict_places[place][i])

# Convert the dictionary to a DataFrame
df_reviews = pd.DataFrame(df_data)

# Example to print the DataFrame
print(df_reviews)
```
